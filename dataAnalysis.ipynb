```python
profile_df = pd.json_normalize(df['profile'])
profile_df = pd.concat(
    [
        df[['candidate_id']].reset_index(drop=True),
        profile_df.reset_index(drop=True)
    ],
    axis=1
)
profile_df
```

```python
redro_df = pd.json_normalize(df['redrob_signals'])
redro_df.sample(20)
```

```python
redro_df.columns
```

```python
rows, col = redro_df.shape
for col in redro_df.columns:
    missing_perc = (redro_df[col].isna().sum()/rows)*100
    if missing_perc > 95:
        redro_df.drop(col, axis=1, inplace=True)
        print(f"Removed {col}")
```
